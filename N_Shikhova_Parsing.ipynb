### Сбор ссылок для парсинга

Переходим на страницы, сформированную с помощью GET-запроса. Листаем эти страницы, собирая на них ссылки, которые ведут на страницы с фильмами.

```python
# Функция для получения ссылок на страницы фильмов

def getLinks(page_number,link_href):

    # составляем ссылку на страницу фильтрации
    if (page_number!=1):
        page_link = link_href+"&start={}&ref_=adv_nxt".format(page_number)
    else:
        page_link = link_href

    # запрашиваем данные по ней
    try:
        response = requests.get(page_link, headers={'User-Agent': UserAgent().chrome})
        html = response.content
        soup = BeautifulSoup(html,'html.parser')
    except:
        print("Не удалось подключиться по html-адресу: "+link_href)
        logger.debug("Не удалось подключиться по html-адресу: "+link_href)
        return []

    if not response.ok:
        # если сервер нам отказал, вернем пустой лист для текущей страницы
        return []



    # наконец, ищем ссылки на фильмы
    
    film_links=[]
    if soup.findAll(lambda tag: tag.name == 'h3' and tag.get('class') == ['lister-item-header'])!=None:
        film_links=soup.findAll(lambda tag: tag.name == 'h3' and tag.get('class') == ['lister-item-header'])

        film_links = ['https://www.imdb.com' + l_k.a.attrs['href'] for l_k in film_links]

    return film_links
```

### Парсинг страницы с фильмом

```python

# Функция для получения информации о фильме

def getFilmInfo(film_link):


    # отправляем запрос
    try:
        response = requests.get(film_link, headers={'User-Agent': UserAgent().chrome})
        html = response.content
        soup = BeautifulSoup(html,'lxml')
    except:
        print("Не удалось подключиться по html-адресу: "+link)
        logger.debug("Не удалось подключиться по html-адресу: "+link)
        return []

    if not response.ok:
        # если сервер нам отказал, вернем пустой лист для текущей страницы
        return []

      
    info=[]

    
    # Название
    
    title=""
    if soup.find('h1')!=None:
        title=soup.find('h1').text.replace("\xa0"," ").strip()
    info.append(title)
    
    #Рейтинг
    
    rating=""
    if soup.find(lambda tag: tag.name == 'div' and tag.get('class') == ['imdbRating'])!=None:
        rating=soup.find(lambda tag: tag.name == 'div' and tag.get('class') == ['imdbRating']).span.text.replace("\xa0"," ").strip()
    info.append(rating)
    
    #Тип (сериал, фильм и т.д.)
    
    type_film=""
    if soup.find("a", attrs={ "title" : "See more release dates"})!=None:
        type_film=soup.find("a", attrs={ "title" : "See more release dates"}).text.replace('\n','').replace("\xa0"," ").strip()
    
        type_film = type_film.split(' ')


        if (type_film[-1].isdigit()):
            type_film_list=type_film[:len(type_film)-3]
            type_film = ' '.join(type_film_list)
        else:
            type_film = ' '.join(type_film)
            if type_film.rfind(')')==len(type_film)-1:
                l = type_film.rfind('(')
                r = type_film.rfind(')')
                type_film = type_film[:l] + type_film[r+1:-1] 
     
    info.append(type_film.strip())
    
    #Жанр(ы) 
    
    genres_film_block=[] 
    infoG=[]
    gG=[] 
    if soup.find_all("div", attrs={ "class" : "see-more inline canwrap"})!=None:
        genres_film_block=soup.find_all("div", attrs={ "class" : "see-more inline canwrap"})
  


        for g in range(len(genres_film_block)):
            text_gG=str(genres_film_block[g])# переводим в текст, чтобы осущ поиск        
            Genres='Genres:' #Заменить на не найденный и проверить на ошибку
            if Genres in text_gG:
            
                gG=genres_film_block[g].find_all("a")
            for g4 in range(len(gG)): # цикл по ссылкам найденным в этом блок
                    gg1=gG[g4].text.strip()
                    gg1= re.sub(r'\s+', ' ', gg1) # удаляем повторяющиеся пробелы
                    infoG.append(gg1)
    info.append(str(infoG).replace("[","").replace("]","").replace("'","")) 
    
    #Топ каста (stars)
    
    cust=[]
    infoCust=[]
    Cust_list=[]  
    if soup.find_all("div", attrs={ "class" : "credit_summary_item"})!=None:
        cust=soup.find_all("div", attrs={ "class" : "credit_summary_item"})
   


        for c_l in range(len(cust)):
            text_c_l=str(cust[c_l])# переводим в текст, чтобы осущ поиск        
            Stars='Stars:' #Заменить на не найденный и проверить на ошибку
            if Stars in text_c_l:
            
                Cust_list=cust[c_l].find_all("a")
            for cs4 in range(len(Cust_list)): # цикл по ссылкам найденным в этом блок
                    st=Cust_list[cs4].text.strip()
                    st= re.sub(r'\s+', ' ', st) # удаляем повторяющиеся пробелы
                    if st!="See full cast & crew":
                        infoCust.append(st)
    info.append(str(infoCust).replace("[","").replace("]","").replace("'","")) 


    #Блок Details
    
    det=[]
    det_info=[]
    det_info2=[]
    if soup.find("div", attrs={ "id" : "titleDetails"})!=None:
        det=soup.find("div", attrs={ "id" : "titleDetails"}).text.replace("\xa0"," ").strip()
    
        det = det.split('\n\n\n')
        for d_l in range(len(det)):
            text_d_l=str(det[d_l])# переводим в текст, чтобы осущ поиск              
            End='Company Credits'
            if End in text_d_l:
                det_info=det[:d_l] # оставляем только нужный блок
                for d_l2 in range (len(det_info)): # чистим код
                    det_info2.append(det_info[d_l2].replace("Edit"," ").replace("Details"," ").replace("See more »"," ").replace("\n"," ").strip())
               
          
    info.append(str(det_info2).replace("[","").replace("]","").replace("'",""))    
    
  
    #Box office
    
    box=""
    if soup.find("a", attrs={ "class" : "pro_title_href"})!=None:
        box=soup.find("a", attrs={ "class" : "pro_title_href"}).attrs['href']
    
    info.append(box) 
    
    #Technical specs
    
    tech=[]
    tech_info=[]
    tech_info2=[]
    if soup.find("div", attrs={ "id" : "titleDetails"})!=None:
        tech=soup.find("div", attrs={ "id" : "titleDetails"}).text.replace("\xa0"," ").replace("\n\n\n","\n\n").strip()
    
    
        tech = tech.split('\n\n')
        for t_l in range(len(tech)):
            text_t_l=str(tech[t_l])# переводим в текст, чтобы осущ поиск        
         
            Technical='Technical Specs' #Заменить на не найденный и проверить на ошибку
            if Technical in text_t_l:
                tech_info=tech[t_l:] # оставляем только нужный блок
                for t_l2 in range (len(tech_info)): # чистим код
                    t_i=tech_info[t_l2].replace("Technical Specs","").replace("See full technical specs »","").replace("Show more on","").replace("IMDbPro »","").replace("\n"," ").strip()
                    if (t_i!=""):                        
                        tech_info2.append(t_i)
               
         
    info.append(str(tech_info2).replace("[","").replace("]","").replace("'",""))    
     
    
    return info
```

### Информация о количестве фильмов для парсинга

Для начала необходимо узнать количество фильмов доступных для парсинга, после заполнения запроса на фильтрацию.

```python
# Функция для получения информации о количестве фильмов

def getCountFilms(link):


    # отправляем запрос
    try:
        response = requests.get(link, headers={'User-Agent': UserAgent().chrome})
        html = response.content
        soup = BeautifulSoup(html,'lxml')
    except:
        print("Не удалось подключиться по html-адресу: "+link)
        logger.debug("Не удалось подключиться по html-адресу: "+link)
        return []

    if not response.ok:
        # если сервер нам отказал, вернем пустой лист для текущей страницы
        return []


    count_films=""
    if ((soup.find("div", attrs={ "class" : "desc"}))!=None):
        count_films=soup.find("div", attrs={ "class" : "desc"}).find("span").text
    
    
        if count_films=="No results." or count_films=="":
            count_films="0"
        else:    
            l = count_films.find('f')
            count_films = count_films[l+1:].replace("titles."," ").strip()
   
    return int(count_films)
```

### Запускаем цикл, который парсит информацию о фильмах и добавляет ее в датафрейм

```python
import pandas as pd
from tqdm import tqdm_notebook
```

```python
# Установка количественного лимита для пагсинга

if getCountFilms(link)!=0: # Цикл парсинга запустится только если будут найдены фильмы, по сгенерированнному запросу фильтрации
    logger.debug("Запуск цикла парсера!")  
    
    # т.к.Гет-запрос вида: start=51,101,...451,501
    if (getCountFilms(link)>1000): # ограничение на парсинг 1000 фильмов, на стр по 50, значит последняя начинается с 951
        limit=1000    
    else:   
        limit=getCountFilms(link)
        
    logger.debug('Определен лимит для парсинга: ',limit) 
    
# Cоздаем датафрейм для записи полученых значений

    info_films_list=[]


    # Цикл парсинга
    for page_number in tqdm_notebook(range(1,limit,50), desc='Pages'):
    # собрали хрефы с текущей страницы списка по фильтрации        
        F_links = getLinks(page_number,link) 
        
        for info in tqdm_notebook(F_links, desc='Films', leave=False):
        # иногда с первого раза страничка не парсится
            for i in range(5):
                try:
                    # запрос информации по фильму
                    data_row = getFilmInfo(info)      
                    # и добавляем её в список
                    info_films_list.append(data_row)
                    # если всё получилось - выходим из внутреннего цикла
                    break
                except:
                    # Иначе, пробуем еще несколько раз, пока не закончатся попытки
                    print('Внимание! Попытка парсинга не удалась: '+info)
                    logger.debug('Внимание! Попытка парсинга не удалась: '+info)
                    continue
else:
    print('Фильмов по введенным параметнам не найдено, парсиг не возможен. Сформируйте новый зпрос на фильтрацию с другими параметрами')
    logger.debug('Фильмов по введенным параметнам не найдено, парсиг не возможен. Сформируйте новый зпрос на фильтрацию с другими параметрами')
```
